## Model Comparison

```python
import time
import pandas as pd
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import accuracy_score

# Your refined categories (exactly as in your example)
categories = {
    "Software Development": [
        "software engineer", "software developer", "junior software developer",
        "associate software engineer", "intern software engineer", "trainee software engineer",
        "web developer", "developer", "devops", "qa engineer", "quality assurance",
        "quality assurance engineer", "test engineer", "test analyst", "ui", "ux", "ui ux"
    ],
    "Data & Analytics": [
        "data analyst", "data engineer", "business analyst",
        "deputy analyst", "senior operations analyst", "analyst",
        "associate business analysis", "machine learning"
    ],
    "Engineering": [
        "engineer", "engineering", "technical officer",
        "infrastructure network engineer", "industrial engineering",
        "associate system engineer", "associate electronics engineer",
        "assistant engineer", "site engineer", "electronics engineer",
        "electronic engineer", "product data management engineer",
        "work study officer", "engineering executive", "project coordinator site engineer"
    ],
    "Marketing & Sales": [
        "marketing", "sales", "seo", "merchandising", 
        "merchandising assistant", "front end product development merchandising executive",
        "sales admin support", "sales administration support",
        "customer engagement specialist"
    ],
    "Human Resources": [
        "management trainee", "executive hr",
        "human resources executive", "hr"
    ],
    "Education & Research": [
        "tutor", "lecturer", "assistant lecturer",
        "demonstrator", "temporary demonstration", "teacher",
        "research assistant", "instructor"
    ],
    "Operations & Support": [
        "operations", "operations assistant manager",
        "customer support", "support engineer",
        "associate application support engineer",
        "coordinator", "planning coordinator", "project executive",
        "planning executive", "planning officer", "assistant planning",
        "executive planning", "production coordinator",
        "production executive", "lean executive",
        "assistant administration executive", "staff assistant",
        "trainee staff assistant", "management assistant",
        "executive banking operation", "compliance officer",
        "executive process improvements", "executive", "operations analyst"
    ],
    "Finance & Accounting": [
        "finance", "accountant executive", "account", "inspector of customs",
        "purchasing officer"
    ]
}

# Example ground truth data for benchmarking
df = pd.DataFrame({
    'Job title': [
        "software engineer", "data analyst", "lecturer",
        "sales executive", "operations manager", "finance officer"
    ],
    'True category': [
        "Software Development", "Data & Analytics", "Education & Research",
        "Marketing & Sales", "Operations & Support", "Finance & Accounting"
    ]
})

# Models to compare
models_to_test = [
    'all-MiniLM-L6-v2',
    'all-mpnet-base-v2',
    'paraphrase-MiniLM-L12-v2',
    'distilbert-base-nli-stsb-mean-tokens'
]

results = []

for model_name in models_to_test:
    print(f"\nTesting model: {model_name}")
    model = SentenceTransformer(model_name)

    # Step 3: Compute category embeddings
    category_embeddings = {
        category: model.encode(keywords) for category, keywords in categories.items()
    }

    # Step 4: Function to classify
    def classify_job_title(job_title):
        job_embedding = model.encode(job_title)
        best_category = None
        max_similarity = -1
        for category, embeddings in category_embeddings.items():
            similarity = max(util.cos_sim(job_embedding, emb).item() for emb in embeddings)
            if similarity > max_similarity:
                max_similarity = similarity
                best_category = category
        return best_category

    # Step 5: Measure time + predict
    start_time = time.time()
    df['Predicted'] = df['Job title'].map(classify_job_title)
    elapsed_time = (time.time() - start_time) / len(df)  # per job title

    # Step 6: Accuracy
    acc = accuracy_score(df['True category'], df['Predicted'])

    # Step 7: Store results
    results.append({
        "Model": model_name,
        "Accuracy": round(acc, 3),
        "Avg Inference Time (s)": round(elapsed_time, 4),
        "Embedding Dim": model.get_sentence_embedding_dimension()
    })

# Results table
results_df = pd.DataFrame(results)
print("\nBenchmark Results:")
print(results_df)

# Save to CSV
results_df.to_csv("model_comparison_results.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

for idx, row in results_df.iterrows():
    plt.scatter(row['Avg Inference Time (s)'], row['Accuracy'], 
                s=row['Embedding Dim']*0.5, label=row['Model'], alpha=0.7)

plt.title('Model Comparison: Accuracy vs. Inference Time')
plt.xlabel('Average Inference Time per Job Title (seconds)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
